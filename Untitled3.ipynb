```python
import pandas as pd
data = {
"Order_ID": [1001, 1002, 1003, 1004],
"Product": ["Laptop", "Phone", "Tablet", "Laptop"],
"Price ($)": [1000, None, 500, 1200],
"Quantity": [1, 2, None, 1],
"Discount (%)": [10, 5, 0, None]
}
```

```python
df = pd.DataFrame(data)
print(df.head(3))
```

```python
print(df.isnull().sum()) # Counts missing values
```

```python
df['Price ($)'].fillna(df['Price ($)'].mean(), inplace=True)
df['Quantity'].fillna(1, inplace=True)
```

```python
df['Revenue'] = (df['Price ($)'] * df['Quantity']) - df['Discount (%)'].fillna(0)
print(df[['Order_ID', 'Revenue']])
```

Draw Pie Chart of given data

```python
import matplotlib.pyplot as plt

# Count occurrences of each product
product_counts = df['Product'].value_counts()

# Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(product_counts, labels=product_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightcoral', 'lightgreen'])
plt.title("Product Distribution in Sales Data")
plt.show()
```
